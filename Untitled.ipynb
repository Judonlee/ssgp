```python
import numpy as np
import gpitch
import gpflow
import matplotlib.pyplot as plt
from gpitch.matern12_spectral_mixture import Matern12sm
```

```python
gpitch.init_settings();
```

```python
Cosine = gpitch.kernels.Cosine
def Kernel_base(input_dim, num_partials, variance, lengthscales, energy, frequencies):
    
    
    kern_exp = gpflow.kernels.Matern12(input_dim=input_dim, variance=variance, lengthscales=lengthscales)

    kern_add = Cosine(input_dim=input_dim, variance=energy[0], frequency=frequencies[0])
                
    for i in range(1, num_partials):
        kern_add += Cosine(input_dim=input_dim, variance=energy[i], frequency=frequencies[i]) 
        
    
    #kern_add.energy = gpflow.param.ParamList([kern_add.kern_list[i].variance for i in range(num_partials)])
    #kern_add.frequencies = gpflow.param.ParamList([kern_add.kern_list[i].frequency for i in range(num_partials)])
    
    final_kern = kern_exp * kern_add
    #final_kern.energy = gpflow.param.ParamList([kern_add.kern_list[i].variance for i in range(num_partials)])
    #final_kern.frequencies = gpflow.param.ParamList([kern_add.kern_list[i].frequency for i in range(num_partials)])

    return final_kern
```

```python
num_partials = 3
energy = np.random.rand(num_partials)
frequency = 10*np.random.rand(num_partials)
var = np.random.rand(1)[0]
leng = np.random.rand(1)[0]
```

```python
frequency
```

```python
kern1 = Matern12sm(input_dim=1, 
                   energy=energy,
                   frequencies=frequency,
                   variance=var,
                   lengthscales=leng)
```

```python
kern2 = Kernel_base(input_dim=1, 
                   num_partials=num_partials,
                   energy=energy,
                   frequencies=frequency,
                   variance=var,
                   lengthscales=leng)
```

```python
x = np.linspace(-1, 1., 1000).reshape(-1, 1)
x0 = np.array(0.).reshape(-1, 1)
```

```python
cov1 = kern1.compute_K(x, x0)
cov2 = kern2.compute_K(x, x0)
```

```python
diag1 = kern1.compute_Kdiag(x)
diag2 = kern2.compute_Kdiag(x)
```

```python
plt.plot(x, diag1, x, diag2, '--');
```

```python
plt.plot(x, cov1, x, cov2, '--');
```

```python
K1 = kern1.compute_K_symm(x)
K2 = kern2.compute_K_symm(x)
```

```python
plt.matshow(K1)
plt.colorbar();

plt.matshow(K2)
plt.colorbar();
```
